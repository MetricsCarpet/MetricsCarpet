```python
import sqlite3
import pandas as pd

def load_measurements(experiments):
    data = {}
    for experiment in experiments:
        name = experiment['name']
        filename = experiment['file']
        con = sqlite3.connect("var/measurements/%s" % experiment['file'])
        data[name] = pd.read_sql("SELECT * from measurements", con).drop('index', axis=1)    
    return data
   
```

```python
infectio_experiments = [
    {
        'name': 'Infectio_v0.1',
        'file': 'Infectio_v0.1_150515003418.sqlite',
    },
    {
        'name': 'Infectio_v0.2',
        'file': 'Infectio_v0.2_150515003418.sqlite',
    },
    {
        'name': 'Infectio_v0.3',
        'file': 'Infectio_v0.3_150515003418.sqlite',
    },
]
    
cp_experiments = [   
    {
        'name': 'CellProfiler_v1.0',
        'file': 'CellProfiler_v1.0_150531230704.sqlite',
    },
]


data = {}
data['infectio'] = load_measurements(infectio_experiments)
data['cp'] = load_measurements(cp_experiments)
```

```python
metric_labels = ['line_counts', 'comments', 'cyclomatic_complexity', 'help_metric']

class MinMaxNorm(object):
    
    def __init__(self):
        self.min = 0
        self.max = 0

    def find_minmax(self, data):
        self.min = min([self.min, data.min()])
        self.max = max([self.max, data.max()])
        
    def normalize_minmax(self, data):
        data_norm = (data - self.min) / (self.max - self.min)
        return data_norm

    def norm_metrics(self, data):
        result = {}
        for name in data:
            result[name] = data[name].copy()
            for metric in metric_labels:
                result[name][metric] = normalize_minmax(result[name][metric])
        return result

    def merge_metrics(self, data):
        for name in data:
            norm_df = data[name].copy().sort('filename').filter(metric_labels, axis=1)
            for metric in metric_labels:
                norm_df[metric] = normalize_minmax(norm_df[metric])

# print data.keys()
                
# Share min-max per column (key)
norm_data = {}
for key in metric_labels:
    normalizer = MinMaxNorm()
    for product in data:
        for version in data[product]:
            normalizer.find_minmax(data[product][version][key])
    for product in data:
        if product not in norm_data:
            norm_data[product] = {}
        for version in data[product]:
            if version not in norm_data[product]:
                norm_data[product][version] = data[product][version].copy()
            norm_data[product][version][key] = normalizer.normalize_minmax(norm_data[product][version][key])

# Merge infectio
infectio_mdata = None
for version in norm_data['infectio']:
    dash = version.replace('Infectio_', '')
    norm_data['infectio'][version] = norm_data['infectio'][version].drop('pathname', axis=1)
    names = dict((n, n + '_' + dash) for n in norm_data['infectio'][version].columns if n != 'filename')
    norm_data['infectio'][version].rename(columns=names, inplace=True)
    if infectio_mdata is None:
        infectio_mdata = norm_data['infectio'][version]
    else:
        infectio_mdata = infectio_mdata.merge(norm_data['infectio'][version], on='filename', how='outer')
# infectio_mdata = norm_data['infectio']['Infectio_v0.1'].merge(norm_data['infectio']['Infectio_v0.2'], on='filename', how='outer').merge(norm_data['infectio']['Infectio_v0.3'], on='filename', how='outer').sort('filename')
infectio_filenames = infectio_mdata['filename']
infectio_mdata = infectio_mdata.drop('filename', axis=1).fillna(0)

# Clean cp
version = 'CellProfiler_v1.0'
dash = 'v1.0'
names = dict((n, n + '_' + dash) for n in norm_data['cp'][version].columns if (n != 'filename' and n != 'pathname'))
norm_data['cp'][version].rename(columns=names, inplace=True)
cp_mdata = norm_data['cp'][version].sort('filename')
cp_filenames = cp_mdata['filename']
cp_mdata = cp_mdata.drop(['filename', 'pathname'], axis=1)

            
# Normalize infectio
# norm_data = norm_metrics(data)
# for name in norm_data:
#     dash = name.replace('Infectio_', '')
#     norm_data[name] = norm_data[name].drop('pathname', axis=1)
#     names = dict((n, n + '_' + dash) for n in norm_data[name].columns if n != 'filename')
#     norm_data[name].rename(columns=names, inplace=True)

# hardcoded merge :)
# mdata = norm_data['Infectio_v0.1'].merge(norm_data['Infectio_v0.2']).merge(norm_data['Infectio_v0.3']).sort('filename')
# filenames = mdata['filename']
# mdata = mdata.drop('filename', axis=1)

# infectio_mdata
# cp_mdata
```

```python
%matplotlib inline
# config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt

import pandas as pd
import numpy as np


import os


if not os.path.exists('imgs'):
    os.makedirs('imgs')

    
def plot_norm_metrics(norm_metrics, size, column_names, row_names, fig_filename=None):
    # Expect `norm_metrics` to contain all the min-max normalized metrics sorted by `filename`


    # Plot it out
    fig, ax = plt.subplots()
    heatmap = ax.pcolor(norm_metrics, cmap=plt.cm.Blues, alpha=0.8)

    ##################################################
    ## FORMAT ##
    ##################################################

    fig = plt.gcf()
    fig.set_size_inches(size[0], size[1], dpi=300)

    # turn off the frame
    ax.set_frame_on(False)

    # put the major ticks at the middle of each cell
    ax.set_yticks(np.arange(norm_metrics.shape[0]) + 0.5, minor=False)
    ax.set_xticks(np.arange(norm_metrics.shape[1]) + 0.5, minor=False)

    # want a more natural, table-like display
    ax.invert_yaxis()
    ax.xaxis.tick_top()

    # Set the labels

    # 
    ax.set_xticklabels(column_names, minor=False) 
    #ax.set_yticklabels(norm_metrics.index, minor=False)
    ax.set_yticklabels(row_names, minor=False)

    # rotate the 
    plt.xticks(rotation=90)
    
    ax.grid(False)

    # Turn off all the ticks
    ax = plt.gca()

    for t in ax.xaxis.get_major_ticks(): 
        t.tick1On = False 
        t.tick2On = False 
    for t in ax.yaxis.get_major_ticks(): 
        t.tick1On = False 
        t.tick2On = False 
    
    if fig_filename is not None:
        plt.savefig(fig_filename, format='svg')
    
        
        
```

```python
'''
Make a colorbar as a separate figure.
'''
def save_bar():
    from matplotlib import pyplot
    import matplotlib as mpl

    # Make a figure and axes with dimensions as desired.
    fig = pyplot.figure(figsize=(0.2,16))
    ax1 = fig.add_axes([0.05, 0.80, 0.9, 0.15])

    # Set the colormap and norm to correspond to the data for which
    # the colorbar will be used.
    cmap = mpl.cm.Blues
    norm = mpl.colors.Normalize(vmin=0, vmax=1)

    # ColorbarBase derives from ScalarMappable and puts a colorbar
    # in a specified axes, so it has everything needed for a
    # standalone colorbar.  There are many more kwargs, but the
    # following gives a basic continuous colorbar with ticks
    # and labels.
    cb1 = mpl.colorbar.ColorbarBase(ax1, cmap=cmap,
                                       norm=norm,
                                       orientation='vertical')
    cb1.set_label('')
    
    plt.savefig('imgs/colorbar.svg', format='svg')

save_bar()
    
```

```python
def get_complexity_score(metrics, versions=['0.1', '0.2', '0.3']):
    cscore = {}
    for version in versions:        
        # cscore[version] = metrics['line_counts_v%s' % version].mean() / metrics['comments_v%s' % version].mean() * metrics['cyclomatic_complexity_v%s' % version].mean()
        #cscore[version] = ((metrics['line_counts_v%s' % version] + (1 - metrics['comments_v%s' % version]) + metrics['cyclomatic_complexity_v%s' % version]) / 3).mean()
        cscore[version] = ((metrics['line_counts_v%s' % version] + metrics['comments_v%s' % version] + metrics['cyclomatic_complexity_v%s' % version]) / 3).mean()
    
    return cscore

def get_norm_filenumber(data, cscore, version_ptrn='Infectio_v%s'):
    filecounts = [data[version_ptrn % key].shape[0] for key in cscore]
    filenum = {}
    for key in cscore:
        filenum[key] = (float(data[version_ptrn % key].shape[0]) - min(filecounts)) / (max(filecounts) - min(filecounts))
    return filenum

def normalize_by_filenumber(data, cscore):
    filenum = get_norm_filenumber(data, cscore)
    ncscore = {}
    for key in cscore:
        ncscore[key] =  (filenum[key] + cscore[key])/2
    return ncscore

infectio_versions = ['0.1', '0.2', '0.3']
infectio_cscore = get_complexity_score(infectio_mdata)
infectio_filenum = get_norm_filenumber(data['infectio'], infectio_cscore)
infectio_ncscore = normalize_by_filenumber(data['infectio'], infectio_cscore)
print 'Infectio scoring:'
for version in infectio_versions:
    print '%s => version score: %s | avaraged by number of files: %s' % (version, infectio_cscore[version], infectio_ncscore[version])

#pd.DataFrame(, columns=['ncomplexity', 'nfilecount'])
```
